## Import libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

## Creating dataset 

```python
n_samples = 1500
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)
```

```python
X
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1])
plt.title("Dataset with Equal Variance")
plt.show()
```

## Incorrect number of clusters

```python
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)
```

```python
y_pred
```

```python
y
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1])
plt.title("Dataset Scatter Diagram")

plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("Incorrect Number of Clusters")
```

## Dataset with different variance

```python
X_varied, y_varied = make_blobs(n_samples=n_samples,
                                cluster_std=[1.0, 1.5, 0.5],
                                random_state=random_state)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X_varied[:, 0], X_varied[:, 1])
plt.title("Dataset with Unequal Variance")
plt.show()
```

```python
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X_varied)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.scatter(X_varied[:, 0], X_varied[:, 1])
plt.title("Dataset with Unequal Variance")

plt.subplot(122)
plt.scatter(X_varied[:, 0], X_varied[:, 1], c=y_pred)
plt.title("Cluster with Unequal Variance")
```

## Dataset with unevenly sized blobs

```python
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X_filtered[:, 0], X_filtered[:, 1])
plt.title("Dataset with Unevenly Sized Blobs")
plt.show()
```

```python
y_pred = KMeans(n_clusters=3,
                random_state=random_state).fit_predict(X_filtered)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1])
plt.title("Dataset with Unevenly Sized Blobs")

plt.subplot(122)
plt.scatter(X_filtered[:, 0], X_filtered[:, 1], c=y_pred)
plt.title("Cluster with Unevenly Sized Blobs")
```

## Dataset with Moon Shape

```python
from sklearn.datasets import make_moons
X_moon, y_moon = make_moons(1500, noise=.05, random_state=170)
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(X_moon[:, 0], X_moon[:, 1])
plt.title("Dataset with Moon Shape")
plt.show()
```

```python
y_pred = KMeans(n_clusters=2, random_state=0).fit_predict(X_moon)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.scatter(X_moon[:, 0], X_moon[:, 1])
plt.title("Dataset with Moon Shape")

plt.subplot(122)
plt.scatter(X_moon[:, 0], X_moon[:, 1], c=y_pred)
plt.title("Cluster with Moon Shapes")
```

```python
from sklearn.cluster import SpectralClustering
y_pred = SpectralClustering(n_clusters=2, affinity='nearest_neighbors').fit_predict(X_moon)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(121)
plt.scatter(X_moon[:, 0], X_moon[:, 1])
plt.title("Dataset with Moon Shape")

plt.subplot(122)
plt.scatter(X_moon[:, 0], X_moon[:, 1], c=y_pred)
plt.title("Cluster with Moon Shapes")
```
